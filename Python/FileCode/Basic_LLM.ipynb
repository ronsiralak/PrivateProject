```python
from transformers import BartForConditionalGeneration, BartTokenizer
```

```python
# Load the pre-trained BART model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)
```

```python
# Input text that you want to summarize
input_text = """
Replace this with your input text that you want to summarize.
This could be a long article or document that you want to condense into a summary.
"""
```

```python
# Tokenize and encode the input text
input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=1024, truncation=True)

# Generate the summary
summary_ids = model.generate(input_ids, max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
```

```python
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Generated Summary:")
print(summary)
```

```python
import requests
from bs4 import BeautifulSoup
from transformers import BartForConditionalGeneration, BartTokenizer
```

```python
# URL of the online essay you want to summarize
url = "https://www.projectpro.io/article/llm-project-ideas/881#mcetoc_1h7ag64apm"

# Send an HTTP GET request to the URL and parse the HTML content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract the text content from the HTML (modify this based on the structure of the webpage)
text_content = soup.get_text()
```

```python
text_content
```

```python

# Tokenize and encode the text content
input_ids = tokenizer.encode(text_content, return_tensors="pt", max_length=1024, truncation=True)

# Generate the summary
summary_ids = model.generate(input_ids, max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)

# Decode and print the summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Generated Summary:")
print(summary)
```
